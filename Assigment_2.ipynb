### *__why scaling is required before PCA__*

#### Scaling is necessary because PCA identifies patterns by maximizing variance, making it highly sensitive to the raw numerical range of the data. For example, in a dataset containing Age (0â€“100) and Income (0â€“100k), a variance of 50 in Age is significant (covering half the range), whereas a variance of 50 in Income is negligible. Without scaling, PCA would mathematically perceive Income as more "important" simply because its numbers are larger, allowing it to dominate the principal components and lead to a biased result. 
#### By scaling the data to a 0 mean and 1 standard deviation, we normalize the features so they contribute equally, ensuring the PCA results reflect the true relationships in the data regardless of the original units

```python
# PCA on feature
from sklearn.preprocessing import StandardScaler # importing library to perform center and scaling
sc = StandardScaler()

dataset = housing.drop(['median_house_value','ocean_proximity'],axis=1)  # drop target value(house value) and string value(ocean proximity)
dataset.fillna(dataset['total_bedrooms'].mean(),inplace=True)  # drop the row with NAN value
data = sc.fit_transform(dataset)

# apply PCA
from sklearn.decomposition import PCA
pca = PCA()
pca.fit_transform(data)
```

```python
explained_variance = (pca.explained_variance_ratio_)*100  # convert explained variance to perrcentage
explained_variance
```

```python
# plot PCA vs variance they explained plot

# PCA labels
pca_labels = [f'PCA{i}' for i in range(1, 9)]

# Create plot
plt.figure(figsize=(8, 5))
plt.bar( pca_labels, explained_variance, color='#ADD8E6',edgecolor='black')

# Labels and title
plt.xlabel('Principal Components', color='#8B0000', fontsize=11)
plt.ylabel('Explained Variance (%)', color='#8B0000', fontsize=11)
plt.title('PCAs vs Variance They Explained', color='#8B0000', fontsize=13)

# Tick color
plt.xticks(color='#8B0000')
plt.yticks(color='#8B0000')

# annotate values on bars
for i, val in enumerate(explained_variance):
    plt.text(i, val+1, f'{val:.2f}%' , horizontalalignment='center',verticalalignment='center', color='#8B0000')
    # first two arguments coordinates i.e tex at (i, val+1)
    # third argument is text
    #all other are text properties

plt.tight_layout()
plt.show()
```

## Implement multiple linear regression 

```python
class multi_linear_reg:
    def __init__(self,learning_rate=0.01,max_iteration=1000):
        self.lr=learning_rate 
        self.a=None # weight
        # take n+1 perameter into weigth last one  is for bias
        
        self.max_iteration=max_iteration
        print('learning rate(lr)=',self.lr,'\nmaximum iteration=', self.max_iteration)
     
    def loss(self,x,y):
        # assume initial line as y=[0]*x + [0]
        loss_fun= ( y - x@self.a )**2
        return np.sum(loss_fun)/(2*len(y))  # no of data points is same as len of y

    # Gradient Decent algo.
    def fit(self,x,y):
        self.raw_x=x
        x = np.array(self.raw_x, dtype=float)
        y = np.array(y, dtype=float).reshape(-1, 1)
     
        # Feature scaling 
        self.x_mean = x.mean(axis=0)
        self.x_std = x.std(axis=0) 
        x = (x - self.x_mean) / self.x_std
     
        m, n = x.shape
     
        # Add bias column
        ones = np.ones((m, 1))
        x = np.hstack((x, ones))  # (m, n+1)
     
        self.x = x
        self.y = y
     
        # Initialize weights
        self.a = np.zeros((n+1, 1))
     
        self.iteration = 0

        print('Print loss function after 100 iteration')
        
        while 1:      
           pre_loss=self.loss(x,y)
           
           self.d_w=0
           self.d_w = (-1/m)*( (x.T)@(y - x@self.a) )  # derivative of loss (n+1,1 )
            
           #define new weights
           self.a -= self.lr*(self.d_w)

           curr_loss=self.loss(x,y)
            
            # condition when gradient decent should stop
           if  abs(pre_loss - curr_loss) < 1e-6 or self.iteration >= self.max_iteration:
               break
           #increse no. of iteration each time by 1       
           self.iteration+=1
           if self.iteration % 100 == 0:
               print(" MSE loss = ",curr_loss)
    
    # calculating R^2 
    # R^2 = (SS(mean) - SS(predicted value)) / (SS(mean))  --> SS()=sum of square residue around  
    def r_square(self):
        y_predicted = self.predict(self.raw_x)

        y_mean = np.repeat(np.mean(self.y), len(self.y))  #np.repeat(number_to_repeat, n_times)
        y_mean=y_mean[:,None]
        SS_mean=sum( (y_mean -self.y)**2 )
        SS_predicted_value = sum( (y_predicted - self.y)**2 )

        return ((SS_mean - SS_predicted_value)/(SS_mean))[0]

    def predict(self, x):
        
       x = np.array(x, dtype=float)
       x = (x - self.x_mean) / self.x_std
    
       m = x.shape[0]
       ones = np.ones((m, 1))
       x = np.hstack((x, ones))
    
       return x @ self.a
    
```

### *__Is high R2 score always good?__*

#### A higher RÂ² value is not always better. In multiple linear regression, adding more independent variables will always increase (or at least not decrease) RÂ², regardless of whether the new features are truly relevant or completely independent of the target. Even irrelevant features can randomly explain a small portion of the variance, causing RÂ² to increase.
#### To address this issue, Adjusted RÂ² is used. Adjusted RÂ² penalizes the model for adding unnecessary features and takes into account the number of predictors, increasing only when a new feature genuinely improves the model.

##### 

### *__Is low training loss always preferred?__*

#### No
#### __Memorization (Overfitting):__ If the loss is too low, the model might just be memorizing the training data. Itâ€™s like a student who memorizes the exact answers to a practice test but fails the real exam because they don't actually understand the subject.
#### __The Size of the Numbers:__ If we are predicting house prices, the numbers are huge. If a house costs \\$1,000,000 and our model predicts      \\$1,010,000 ,our error is only 1%. Even though the model is very accurate, the "Loss" (the math difference) will look like a huge, scary number (\\$10,000). High loss doesn't always mean a bad model; it just means we are dealing with big targets
